Asistente Inteligente para Planificacion de Alimentacion Saludable

**PROBLEMATICA**

Este proyecto propone una solución basada en IA para apoyar a las personas en la planificación de dietas saludables y personalizadas.
Mediante el uso de modelos de lenguaje (texto→texto) y generadores de imágenes (texto→imagen), se busca brindar planes de alimentación claros, adaptados a distintas necesidades y acompañados de visualizaciones atractivas que refuercen la adherencia a hábitos nutricionales.
Actualmente, las personas se encuentran con una sobrecarga de información demasiado genérica y contradictoria o poco especifica. La planificación de comidas, el cálculo de macronutrientes y la creación de listas de compras consumen una cantidad significativa de tiempo y esfuerzo, lo que lleva a la frustración y al abandono de las metas. La falta de una visualización atractiva de los platillos hace que las personas no se sientan motivadas a seguir un plan. Por lo que, parece importante desarrollar una solución, ya que una nutrición adecuada es fundamental para la salud y el bienestar, y facilitar este proceso puede tener un impacto directo y positivo en la vida de las personas.

**DESARROLLO**

Desarrollo de la Propuesta
•	La solución propuesta es utilizar la inteligencia artificial generativa de Gemini para crear un plan de alimentación personalizado y completo de 7 días, enfocado en la pérdida de grasa y el aumento de masa muscular. El proyecto se vincula directamente con el desarrollo de modelos de IA, ya que se implementará la generación de prompts en los dos modelos vistos en el curso: texto-texto y texto-imagen.

El proceso se dividirá en dos fases principales de generación mediante prompts:
Fase 1: Generación de Contenido de Texto con Gemini (Modelo texto-texto)
Se crearán prompts específicos para generar todos los componentes textuales del plan de 7 días. El objetivo es obtener respuestas detalladas y estructuradas.

•	Prompt para Plan Semanal General:
•	Ejemplo: "Actúa como un nutricionista experto. Crea un plan de comidas detallado para 7 días para una mujer adulta de 31 años, 58 kg y 1.60 m de altura, que como actividad física realiza 3 dias fuerza, 3 días padel y 1 o dos caminatas semanalmente, cuyo objetivo es perder grasa y aumentar masa muscular. El plan debe incluir desayuno, almuerzo, merienda, cena y un snacks por día. Asegúrate de que cada día el total calórico sea de aproximadamente 2200 kcal, con una distribución de macronutrientes de 40% proteínas, 30% carbohidratos y 30% grasas."

•	Prompt para Recetas Detalladas:
•	Ejemplo: "Basado en el [desayuno del Día 1] del plan anterior, genera una receta detallada para 'Tostadas integrales con aguacate y huevo'. La receta debe incluir: una lista de ingredientes con cantidades exactas, instrucciones de preparación paso a paso, el tiempo de preparación estimado y una tabla con el desglose de macronutrientes (proteínas, carbohidratos, grasas) y calorías."

•	Prompt para Lista de Compras Semanal:
•	Ejemplo: "A partir del plan de comidas completo de 7 días que generaste, crea una lista de compras semanal consolidada. Organiza la lista por categorías de alimentos (ej. Frutas y verduras, carnes y aves, lácteos, despensa) para facilitar las compras."
Fase 2: Generación de Imágenes con Gemini (Modelo texto-imagen)
Para cada una de las recetas generadas, se creará un prompt para visualizar el plato final. Esto hará que el plan sea más atractivo y motivador.

•	Prompt para Generación de Imágenes de Comida:
•	Ejemplo 1 (Almuerzo): "Fotografía de comida hiperrealista, de un 'Wok de pollo y vegetales con arroz integral' en un bol de cerámica blanca. El pollo está dorado, los pimientos rojos, brócoli y zanahorias se ven frescos y crujientes, con un ligero brillo de salsa de soja. La iluminación es natural y brillante, resaltando las texturas. El fondo es una mesa de madera oscura."
•	Ejemplo 2 (Desayuno): "Imagen fotorrealista de un tazón de yogur griego con granola casera, rodajas de fresa fresca y un puñado de arándanos. El yogur tiene una textura cremosa. La foto está tomada con una lente que da una profundidad de campo baja que desenfoca suavemente el fondo."

```python
import ipywidgets as widgets
from IPython.display import display
import google.generativeai as genai
```

```python
model = genai.GenerativeModel("gemini-1.5-flash")
```

```python
# Crear un campo de texto para el prompt
prompt_area = widgets.Textarea(
    value='',
    placeholder='Escribe tu prompt sobre alimentación saludable...',
    description='Prompt:',
    layout=widgets.Layout(width='600px', height='150px')
)

# Botón para enviar el prompt
button = widgets.Button(description="Enviar")

# Función al presionar el botón
def on_button_click(b):
    user_prompt = prompt_area.value
    if user_prompt.strip() == "":
        print(" Por favor ingresa un prompt.")
    else:
        print(" Prompt capturado:", user_prompt)
        # Llamar al modelo Gemini (using gemini-1.5-flash for text generation)
        response = genai.GenerativeModel("gemini-1.5-flash").generate_content(user_prompt)
        print("\n Respuesta del modelo:\n")
        print(response.text)

        # Imprimir estadísticas de tokens dentro de la función
        print("\n--- Estadísticas de tokens ---")
        try:
            print("Tokens de entrada:", response.usage_metadata.prompt_token_count)
            print("Tokens de salida:", response.usage_metadata.candidates_token_count)
            print("Tokens totales:", response.usage_metadata.total_token_count)
        except AttributeError:
            print("No se pudo obtener la información de uso de tokens para esta respuesta.")


# Asociar el botón a la función
button.on_click(on_button_click)

# Mostrar en la celda
display(prompt_area, button)
```

```python
#Visualizacion de un menu generado; plan alimenticio transformado a pandas

import pandas as pd

menu = {
    "Dia": ["Lunes", "Lunes", "Lunes", "Lunes"],
    "Comida": ["Desayuno", "Almuerzo", "Merienda", "Cena"],
    "Plato": [
        "2 huevos revueltos con espinacas + 1 tostada integral con aguacate + 100ml leche vegetal de almendras + 5g creatina",
        "150g Pollo a la plancha + 100g  arroz integral + 100g brócoli al vapor",
        "Batido de proteína (30g proteína en polvo) + 1 plátano",
        "150g  Lentejas estofadas + 50g ensalada mixta con aceite de oliva virgen extra"
    ],
    "Calorias": ["400",
                 "550",
                 "300",
                 "450"]
  }

df_menu = pd.DataFrame(menu)
df_menu
```

```python
#Grafico de barras para visualizar distribucion de calorias

import matplotlib.pyplot as plt

df_menu["Calorias"] = df_menu["Calorias"].astype(int)
df_menu.groupby("Comida")["Calorias"].mean().plot(kind="bar")
plt.title("Distribución calórica por comida")
plt.ylabel("Calorías")
plt.show()
```
